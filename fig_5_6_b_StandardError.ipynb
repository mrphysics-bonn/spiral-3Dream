# Standard error of the estimated linear fit

## Python standard libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Plotting

```python
beta = np.array([5,8,10,15])
std_err = np.array([0.001175,0.001324,0.001460,0.001996])*1e3

fig=plt.figure(figsize=(12, 12))
fig.suptitle('Standard error of estimation', fontsize=50)

plt.plot(beta,std_err,'*-',markersize=40, linewidth=4)
plt.xlabel(r'$\beta$', fontsize=45)
plt.ylabel(r'standard error [$\cdot 10^{3}$]', fontsize=45)
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.tick_params(labelsize=40)
```
